```python
import os
import shutil
import random
```

```python
folder = os.path.dirname(os.path.dirname(os.getcwd()))
folder
```

# Renaming files

```python
directory_path = r''+folder+r'\Datasets\ORIGA\Labels'
filenames = os.listdir(directory_path)
filenames
```

```python
# Define the directory containing the files
# List the files in the directory


# Iterate through the filenames and rename them
for filename in filenames:
    # Split the filename into name and extension
    name, extension = os.path.splitext(filename)

    # Add "x" to the name and keep the extension
    new_name = "x" + name + "x" + extension

    # Construct the old and new file paths
    old_path = os.path.join(directory_path, filename)
    new_path = os.path.join(directory_path, new_name)

    # Rename the file
    os.rename(old_path, new_path)
```

```python
for filename in filenames:
    name, extension = os.path.splitext(filename)
    if name.startswith("x") and name.endswith("x"):
        new_filename = name[1:-1]+ extension  # Remove 'x' prefix and 'xx' suffix
        old_file_path = os.path.join(directory_path, filename)
        new_file_path = os.path.join(directory_path, new_filename)
        os.rename(old_file_path, new_file_path)
```

# Train-Test Split

```python
source_folder = r''+folder+r'\Data'
source_image_folder = r''+folder+r'\Data\ImagesCroppedAugmented'
source_label_folder = r''+folder+r'\Data\LabelsCroppedAugmented'

train_image_folder = source_image_folder + "\Train"
train_label_folder = source_label_folder + "\Train"

valid_image_folder = source_image_folder + "\Valid"
valid_label_folder = source_label_folder + "\Valid"

test_image_folder = source_folder + "\Test"
test_label_folder = source_folder + "\Test_gt"
```

```python
os.makedirs(train_image_folder, exist_ok=True)
os.makedirs(valid_image_folder, exist_ok=True)
os.makedirs(test_image_folder, exist_ok=True)
```

```python
os.makedirs(train_label_folder, exist_ok=True)
os.makedirs(valid_label_folder, exist_ok=True)
os.makedirs(test_label_folder, exist_ok=True)
```

```python
file_image_list = os.listdir(source_image_folder)
file_label_list = os.listdir(source_label_folder)
```

```python
total_files = len(file_image_list)

train_size = int(0.75 * total_files)
valid_size = int((total_files - train_size)/2)
test_size = total_files - train_size - valid_size
```

```python
for file_name in file_image_list[:train_size]:
    src_image_path = os.path.join(source_image_folder, file_name)
    src_label_path = os.path.join(source_label_folder, file_name.replace("jpg", "mat"))
    
    dst_image_path = os.path.join(train_image_folder, file_name)
    dst_label_path = os.path.join(train_label_folder, file_name.replace("jpg", "mat"))
    
    shutil.copy(src_image_path, dst_image_path)
    shutil.copy(src_label_path, dst_label_path)
```

```python
for file_name in file_image_list[train_size:train_size + valid_size]:
    src_image_path = os.path.join(source_image_folder, file_name)
    src_label_path = os.path.join(source_label_folder, file_name.replace("jpg", "mat"))
    
    dst_image_path = os.path.join(valid_image_folder, file_name)
    dst_label_path = os.path.join(valid_label_folder, file_name.replace("jpg", "mat"))
    
    shutil.copy(src_image_path, dst_image_path)
    shutil.copy(src_label_path, dst_label_path)
```

```python
for file_name in file_image_list[train_size + valid_size:]:
    src_image_path = os.path.join(source_image_folder, file_name)
    src_label_path = os.path.join(source_label_folder, file_name.replace("jpg", "mat"))
    
    dst_image_path = os.path.join(test_image_folder, file_name)
    dst_label_path = os.path.join(test_label_folder, file_name.replace("jpg", "mat"))
    
    shutil.copy(src_image_path, dst_image_path)
    shutil.copy(src_label_path, dst_label_path)
```
